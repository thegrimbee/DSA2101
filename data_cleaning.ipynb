```python
import pydytuesday
import pandas as pd
import seaborn as sns
import matplotlib as plt


all_recipes = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-09-16/all_recipes.csv')
cuisines = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-09-16/cuisines.csv')
```

```python

all_recipes.head()
cuisines.head()


```

```python
#Focus on following criterions: country, calories, total_time for cuisines dataset

print(cuisines.info())#print(cuisines.describe())

print(cuisines["country"].isnull().sum())
print(cuisines["calories"].isnull().sum())
print(cuisines["total_time"].isnull().sum())
print(cuisines["avg_rating"].isnull().sum())


#Drop all rows with null entries for calories, avg_rating columns
#Also drop rows where total_time equals 0

clean_cuisines = cuisines.dropna(subset = ["calories", "avg_rating"]).copy()
clean_cuisines = clean_cuisines[~(clean_cuisines['total_time'] == 0)]
print(clean_cuisines.info())

#Change column entries to appropriate data type
clean_cuisines["country"] = clean_cuisines["country"].astype('category')




```

```python
#We want to calculate the average rating of the recipes for each country
cuisines_by_country = clean_cuisines.groupby("country")['avg_rating'].agg('mean').sort_values(ascending = False).reset_index().copy()
cuisines_by_country.head()

cuisines_by_calories = clean_cuisines[["calories", "avg_rating"]].sort_values('calories').reset_index(drop = True).copy()
cuisines_by_calories.head()

cuisines_by_total_time = clean_cuisines[['total_time', 'avg_rating']].sort_values('total_time').reset_index(drop = True).copy()
cuisines_by_total_time.head()
```
